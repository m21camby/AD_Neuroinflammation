This file is for checking HSV organoids pri-miRNAs to see coding and intergenic pri-miRNAs

```python
import pandas as pd
import re
import numpy as np
import json
import pickle
```

```python
df = pd.read_csv("/data/rajewsky/home/skim/HSV_organoids/miRNA_analysis/putative_intergenic_pri_miRNA.txt", header=None, sep = "\t")
```

```python
df
```

```python
df.iloc[0,8]
```

```python
# Extract miRNA ID
gene_id_intergenic = []

for i in range(0,133):
    # extract until ;
    gene_id_1st = re.match(r'[^;]+', df.iloc[:,8][i])
    # remove gene_id text
    gene_id_1st = gene_id_1st.group()
    gene_id_2nd = re.sub(r'gene_id \"', " ", gene_id_1st)
    # remove last double quotes
    gene_id_3rd = re.sub(r'\"', " ", gene_id_2nd)
    # remove white space
    gene_id_4th = re.sub(r' ', "", gene_id_3rd)
    gene_id_intergenic.append(gene_id_4th)
```

```python
gene_id_intergenic
```

```python
intergenic_list = list(set(gene_id_intergenic))
intergenic_list
```

```python
df2 = pd.read_csv("/data/rajewsky/home/skim/HSV_organoids/miRNA_analysis/personalized_hsa_pri_miRNA_add_mir200_unique_sorted.gtf", header=None, sep = "\t")
```

```python
# Extract miRNA ID
gene_id = []

for i in range(0,126):
    # extract until ;
    gene_id_1st = re.match(r'[^;]+', df2.iloc[:,8][i])
    # remove gene_id text
    gene_id_1st = gene_id_1st.group()
    gene_id_2nd = re.sub(r'gene_id \"', " ", gene_id_1st)
    # remove last double quotesimport pickle
    gene_id_3rd = re.sub(r'\"', " ", gene_id_2nd)
    # remove white space
    gene_id_4th = re.sub(r' ', "", gene_id_3rd)
    gene_id.append(gene_id_4th)
```

```python
gene_id
```

```python
whole_list = list(set(gene_id))
whole_list
```

```python
coding_list = list(set(whole_list) - set(intergenic_list))
coding_list
```
