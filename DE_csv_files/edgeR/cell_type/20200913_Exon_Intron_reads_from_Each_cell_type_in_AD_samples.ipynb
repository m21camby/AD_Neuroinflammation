### 7. plot

```python
# normalization
raw_data = {'exon': [586, 136, 34, 13, 45, 21],
            'intron': [819, 153, 36, 23, 45, 26]}
df = pd.DataFrame(raw_data)

# From raw value to percentage
totals = [i+j for i,j in zip(df['exon'], 
df['intron'])]

exon = [i / j * 100 for i,j in zip(df['exon'], totals)]
intron = [i / j * 100 for i,j in zip(df['intron'], totals)]
```

```python
r = [0,1,2,3,4,5]

# plot
barWidth = 0.8
names = ('EX','IN','MG','AS','OL','OPC')
# Create PFHxA Bars
plt.bar(r, exon, color='#A20101', edgecolor='white', width=barWidth, 
label="exon")
# Create PFHpA Bars
plt.bar(r, intron, bottom=exon, color='#76AD3B', edgecolor='white', 
width=barWidth, label="intron")

# Custom x axis
plt.xticks(r, names)
plt.xlabel("Sample ID")

# Custom y axis
plt.ylabel("percentage")

# Add a legend
plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)

plt.rcParams['figure.figsize'] = [10, 7]
plt.rcParams['figure.dpi'] = 100

# Show graphic
plt.show()
```
